# Basic TF-IDF Model to Score the 5 Culture Values

# 01 - Preprocessing Text

```python
def load_and_preprocess_text(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text
```

# # 02 - Calculating Tf-Idf

```python
def calculate_tfidf(text, word_lists):
    # Combine all words from word_lists into a single list
    all_words = list(set([word for sublist in word_lists.values() for word in sublist]))
    # Create a TfidfVectorizer with the vocabulary set to all_words
    vectorizer = TfidfVectorizer(vocabulary=all_words)
    # Fit and transform the text data
    tfidf_matrix = vectorizer.fit_transform([text])
    # Create a DataFrame to store TF-IDF scores
    tfidf_scores = pd.DataFrame(tfidf_matrix.T.toarray(), index=vectorizer.get_feature_names_out(), columns=["tfidf"])
    return tfidf_scores
```

# 03 - Calculating Combined Tf-Idf for each culture value

```python
def main(file_path):
    # Load and preprocess text data
    text = load_and_preprocess_text(file_path)
    
    #Remove  duplicates from word_lists
    for parameter, word_list in word_lists.items():
        word_lists[parameter] = list(set(word_list))
    
    # Calculate TF-IDF scores
    tfidf_scores = calculate_tfidf(text, word_lists)
    
    # Print the combined TF-IDF value for each parameter
    print("Combined TF-IDF value for each parameter:")
    for parameter in word_lists.keys():
        combined_tfidf = tfidf_scores.loc[word_lists[parameter], 'tfidf'].sum()
        print(f"{parameter.capitalize()}: {combined_tfidf:.4f}")
```

# 04 - Using other methods for assigning parameter values

# Preprocessing

```python
# Required Libraries
import numpy as np
import re
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from gensim.models import Word2Vec, LdaModel
from gensim.corpora import Dictionary
import gensim.downloader as api

# Function to load data from a file
def load_data(file_path):
    with open(file_path, 'r') as file:
        documents = file.readlines()
    return documents

# Function to preprocess text
def preprocess_text(text):
    # Basic text preprocessing steps
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Function to print parameter values
def print_parameter_values(parameter, scores):
    print(f"Combined score for {parameter.capitalize()}: {scores:.4f}")

# Function to compute combined score for a parameter
def compute_combined_score(model_scores, word_lists, parameter):
    words = word_lists[parameter]
    score = sum(model_scores.get(word, 0) for word in words)
    return score
```

# LSA

```python
# Example function to apply LSA
def apply_lsa(documents, n_components=5):
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(documents)
    svd = TruncatedSVD(n_components=n_components)
    lsa = svd.fit_transform(X)
    return lsa, svd, vectorizer

# Main function for LSA
def main_lsa(file_path):
    # Load and preprocess text data
    documents = load_data(file_path)
    documents = [preprocess_text(doc) for doc in documents]

    # Example of using LSA
    lsa_results, svd, vectorizer = apply_lsa(documents)

    print('Results using LSA:')
    for parameter in word_lists.keys():
        combined_lsa = compute_combined_score(dict(zip(vectorizer.get_feature_names_out(), svd.components_.sum(axis=0))), word_lists, parameter)
        print_parameter_values(parameter, combined_lsa)

main_lsa(file_path='/kaggle/input/sample-data-transcripts/amazon.txt')
```

# Word2Vec

```python
# Example function to train Word2Vec
def train_word2vec(documents):
    sentences = [doc.split() for doc in documents]
    model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
    return model

# Main function for Word2Vec
def main_word2vec(file_path):
    # Load and preprocess text data
    documents = load_data(file_path)
    documents = [preprocess_text(doc) for doc in documents]

    # Example of training Word2Vec
    word2vec_model = train_word2vec(documents)

    print('Results using Word2Vec:')
    for parameter in word_lists.keys():
        combined_word2vec = compute_combined_score({word: word2vec_model.wv[word].mean() for word in word2vec_model.wv.index_to_key}, word_lists, parameter)
        print_parameter_values(parameter, combined_word2vec)

main_word2vec(file_path='/kaggle/input/sample-data-transcripts/amazon.txt')
```

# GloVe

```python
# Example function to load GloVe
def load_glove():
    model = api.load("glove-wiki-gigaword-100")  # Pre-trained GloVe model
    return model

# Main function for GloVe
def main_glove(file_path):
    # Load and preprocess text data
    documents = load_data(file_path)
    documents = [preprocess_text(doc) for doc in documents]

    # Example of loading GloVe
    glove_model = load_glove()

    print('Results using GloVe:')
    for parameter in word_lists.keys():
        combined_glove = compute_combined_score({word: glove_model[word].mean() if word in glove_model else 0 for word in word_lists[parameter]}, word_lists, parameter)
        print_parameter_values(parameter, combined_glove)

main_glove(file_path='/kaggle/input/sample-data-transcripts/amazon.txt')
```
