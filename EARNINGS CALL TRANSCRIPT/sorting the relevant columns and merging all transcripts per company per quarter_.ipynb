# analyzing the no of unique companiees per quarter in our dataset


```python
import pandas as pd
import plotly.express as px

# Assuming 'df' is your DataFrame

# Step 1: Convert 'q' column to PeriodIndex if it's not already
df['q'] = pd.PeriodIndex(df['q'], freq='Q')

# Step 2: Count unique companies per quarter
unique_companies_per_quarter = df.groupby('q')['ticker'].nunique().reset_index()

# Convert PeriodIndex to string for compatibility with Plotly
unique_companies_per_quarter['q'] = unique_companies_per_quarter['q'].astype(str)

# Rename the columns for clarity
unique_companies_per_quarter.columns = ['Quarter', 'Unique Companies']

# Step 3: Plotting the results
fig = px.bar(unique_companies_per_quarter,
             x='Quarter',
             y='Unique Companies',
             title='Number of Unique Companies per Quarter',
             labels={'Quarter': 'Quarter', 'Unique Companies': 'Count of Unique Companies'},
             color='Unique Companies')

# Customize layout
fig.update_layout(xaxis_title='Quarter', yaxis_title='Unique Companies', xaxis_tickangle=-45)

# Show plot
fig.show()
```

from this plot we can easily verify that 2020-Q4 and 2021-Q1 have the most no of unique companies
Also 2020-Q1 to 2020-Q4 have decent number of companies which is a positive sign as we can analyse the five culturual values and their shift before and after covid-19

```python
import pandas as pd

# Assuming `df` is your existing DataFrame
data=df.copy()

# Sample DataFrame
df_ = pd.DataFrame(data)

# Convert 'q' column to PeriodIndex
df_['q'] = pd.PeriodIndex(df_['q'], freq='Q')

# Extract year from quarter
df_['year'] = df_['q'].dt.year

# Group by company (ticker) and quarter, aggregating questions and answers
result_df = df_.groupby(['ticker', 'q', 'year']).agg({
    'transcript': ' '.join

}).reset_index()

# Rename columns for clarity
result_df.columns = ['company', 'quarter', 'year', 'transcript']

result_df.head()
```
